```python
def format_phone(x):
    if pd.isnull(x):
        return x
    
    x = int(float(x))
    
    if len(str(x)) >= 9:
        return str(x).zfill(10)
    
    else:
        return str(x)

customer_df["Phone"] = customer_df["Phone"].apply(format_phone)
```
